## Figuras Principais E Como Ler


```python
explanations = json.loads((OUT / 'figure_explanations_ptbr.json').read_text(encoding='utf-8'))
for name in ['target_position_effect_map.png', 'quantum_vs_classical_delta_map.png', 'classification_article_panel.png', 'fractal_msd_and_geometry.png', 'research_journey_dashboard.png']:
    path = OUT / 'figures' / name
    if path.exists():
        info = explanations.get(name, {})
        display(Markdown(f"### {info.get('title', name)}\n\n{info.get('axes', '')}\n\n{info.get('reading', '')}"))
        display(Image(filename=str(path)))
```
