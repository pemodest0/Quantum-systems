## Metrics

```python
metrics = json.loads((OUT / 'metrics.json').read_text(encoding='utf-8'))
display(Markdown(f"""
- Verdict: `{metrics.get('scientific_verdict')}`
- Open signatures: `{metrics.get('open_signature_count')}`
- Group split combined accuracy: `{metrics.get('group_combined_accuracy'):.3f}`
- Classical-only accuracy: `{metrics.get('group_classical_accuracy'):.3f}`
- Strongest mean dephasing gain: `{metrics.get('max_gain_mean'):.3f}`
- Numerics pass: `{metrics.get('numerics_pass')}`
"""))
```

## Paper-By-Paper Status

```python
paper_verdicts_path = PAPER_OUT / 'paper_verdicts.json'
if paper_verdicts_path.exists():
    verdicts = json.loads(paper_verdicts_path.read_text(encoding='utf-8'))
    lines = ['| Paper | Verdict | Claims | Confidence |', '|---|---:|---:|---:|']
    for paper_key, payload in sorted(verdicts.items()):
        lines.append(f"| `{paper_key}` | `{payload.get('verdict')}` | {payload.get('claim_count')} | {payload.get('mean_confidence'):.2f} |")
    display(Markdown('\n'.join(lines)))
else:
    display(Markdown('No paper reproduction suite output found yet.'))
```

## Paper Suite Figures

```python
for name in ['paper_verdict_overview.png', 'paper_claim_confidence.png']:
    path = PAPER_OUT / 'figures' / name
    if path.exists():
        display(Markdown(f'### {name}'))
        display(Image(filename=str(path)))
```
